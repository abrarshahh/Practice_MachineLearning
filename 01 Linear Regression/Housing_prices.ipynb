```python
#importing the libararies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Load the data
house_data = pd.read_csv('/content/Housing.csv')
```

```python
#check the data
house_data.head()
```

```python
house_data.describe()
```

```python
#visualize the data - area against the price
plt.scatter(house_data['area'],house_data['price'])
```

```python
#define the input data (x) and define the output / prediction (y)
y = house_data['price']

#drop the columns in input data that are already in y or that are useless
columns_to_drop = ['price', 'mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea', 'furnishingstatus']
x = house_data.drop(columns_to_drop, axis=1)
```

```python
#train the data on a LinearRegression model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
#split the data into training and testing set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 3)
```

```python
#checking the shape of training and test sets

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
#training of model using fit() method

model.fit(x_train, y_train)
```

```python
#lets score the model against test data

model.score(x_test, y_test)
```
